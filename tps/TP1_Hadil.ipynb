```python
eastwood = artists.find_one({"first_name": "Clint", "last_name": "Eastwood"})

print('Films dont le directeur est Clint Eastwood :')

movies_directeur = movies.find({"director._id": eastwood['_id']}, {"title": 1, "director._id" : 1})
for movie in movies_directeur:
    print(movie)
```

```python
print("Films dont le titre commence par 'Re' : ")
# Dans le shell mongo :
# db.movies.find({"title": /^Re/}, {"actors": null, "title": 1}).pretty()
# Resultat obtenu : 
# { "_id" : "movie:36", "title" : "Reservoir dogs", "actors" : null }

movies__title_re = movies.find({"title": { "$regex": '^Re' }}, {"actors": 0, "summary": 0})
for movie in movies_title_re:
    print(movie)
print("\n")
```

```python
print("Films dont le titre fini par 'la' : ")
# Dans le shell mongo :
# db.movies.find({"title": /la$/}, {"actors": null, "title": 1}).pretty()
# Resultat obtenu : 
# { "_id" : "movie:22", "title" : "Godzilla", "actors" : null }

movies_title_la = movies.find({"title": { "$regex": 'la$' }}, {"actors": 0, "summary": 0})
for movie in movies_title_la:
    print(movie)
print("\n")
```
