```python
# Pagination des résultats ?

db = client['movies_artists']
movies = db['movies']

#Q : Afficher les 12 documents (films) à partir du dixième inclus en se limitant aux title et _id ?
movies_conditions = movies.find({}, {"_id": 1, "title": 1}).skip(9).limit(12)
for movie in movies_conditions:
  print(movie)
```

```python
# Trier les resultats ? 

print("Par ordre croissant:")

ascending_order = movies.find({}, {"_id": 1, "title": 1}).skip(9).limit(12).sort("title", 1)
for movie in ascending_order:
  print(movie)

print('\n')

print("Par ordre decroissant:")

descending_order = movies.find({}, {"_id": 1, "title": 1}).skip(9).limit(12).sort("title", -1)
for movie in descending_order:
  print(movie)
```

```python
# Critères de recherche simples ?

# Q : Afficher les infos du film dont l'id est 2 ? Se limiter aux infos relatives au titre, au genre et au summary ?
print("Film dont l'id est 2 :")

movie_id_2 = movies.find_one({"_id": "movie:2"}, {"title" : 1, "genre" : 1, "summary" : 1})
print(movie_id_2)

print('\n')

# Q : Combien de films ont été produit en en 1979 et les afficher en se limitant aux infos relatives au titre, à l'année et à la liste des acteurs suivant un ordre alphabétique décroissant sur le titre.
print("Films et nombre de films qui ont été produit en en 1979 :")

movies_1979 = movies.find({"year": 1979}, {"title" : 1, "year" : 1, "actors" : 1}).sort("title", -1)
count = 0
for movie in movies_1979:
    count += 1
    print(movie)
print(count)

print('\n')

# Q : Afficher les infos sur le film dont le genre est "Science-fiction" et produit en 1979 sans préciser l'_id et le summary ?
print('Film dont le genre est "Science-fiction" et est produit en 1979 :')

movies_1979_sf = movies.find({"year": 1979, "genre": "Science-fiction"}, {"title" : 1})
for movie in movies_1979_sf:
    print(movie)
```

```python
# Critères de recherche complexes ?

# Q : Combien de films produits par le director dont l'id est 4 et les afficher de plus récent au moins récent (sans le résumé ni le country ni le genre) ?
print("Films et nombre de films par le director dont l'id est 4 : ")

count = 0
movies_director_4 = movies.find({"director": {"_id" : "artist:4"}}, {"summary" : 0, "country" : 0, "genre" : 0}).sort("year", -1)
for movie in movies_director_4:
    count += 1
    print(movie)
print(count)

print('\n')

# Q : Combien de films produits par le director qui porte l'id 4 et dont l'un des acteurs a joué le rôle de Maximus et les afficher sans le résumé ni le country ni le genre ?
print("Films et nombre de films produits par le director qui porte l'id 4 et dont l'un des acteurs a joué le rôle de Maximus :")

movies_director_4_role = movies.find({"director._id": "artist:4", "actors.role": "Maximus"}, {"summary" : 0, "genre": 0, "country": 0})
count = 0
for movie in movies_director_4_role:
    count += 1
    print(movie)
print(count)
print('\n')

# Q : Ecrire la dernière Qry en déclarant des variables ?

print("Résultat avec variables :")

count = 0
director_id = "artist:4"
role = "Maximus"

movies_director_role_variables = movies.find({"director._id": director_id, "actors.role": role}, {"summary" : 0, "genre": 0, "country": 0})
for movie in movies_director_role_variables:
    count += 1
    # print(movie)

print(count)
```

```python
# Critère de recherche par intervalles 

# Q : Ts les films produits après 2000 compris et avant 2005 exlus (dans un ordre chronologique) en se limitant au titre du film et son année de production ?

print("Films produits après 2000 compris et avant 2005 :")

movies_2000_2005 = movies.find({"year":{"$gte":2000,"$lt":2005}}, {"title":1, "year": 1}).sort("year", 1)
for movie in movies_2000_2005:
    print(movie)
```

```python
# Opérateurs booléens 

# Q : Lister les films parus en 1997 ou avec l'acteur id = 147 ?

print('Films parus en 1997 :')
movies_1997 = movies.find({"$or" : [{"year": 1997}, {"actors._id": "artist:147"}] })
for movie in movies_1997:
    print(movie)
    print('\n')
```

```python
# Opérateurs ensemblistes : in, not in, any, all, exists 

# Q : Combien de films dans lesquels ont joué au moins un des artistes suivants : 34, 98 et 1 et afficher leur titre et la liste des actors ; sens du tri : chronologique et alphabétique croissant sur les noms des artistes ?

print("Films et nombre de films dans lesquels ont joué au moins un des artistes suivants : 34, 98 et 1 : \n")

movies_conditions = movies.find({"actors._id": {"$in": ["artist:34", "artist:98", "artist:1"]}}, {"title": 1, "actors": 1}).sort("actors", 1)
count = 0
for movie in movies_conditions:
    count += 1
    print(movie)

print("\n")
print(count)

print("\n")
    
# Q : Chercher les films dans lesquels ont joué ensemble les actors 23 et 147 ?

print("Films dans lesquels ont joué ensemble les actors 23 et 147 \n")
actors_23_147 = movies.find({"actors._id": {"$all": ["artist:23", "artist:147"]}}, {"title": 1, "actors": 1})
for movie in actors_23_147:
    print(movie)

print("\n")

# Q : Combien de films dans lesquels ont joué ni l'acteur 34, ni 98, ni 1 et afficher leur titre et la liste des actors ; sens du tri : chronologique décroissant et alphabétique croissant sur les titres des films ?

print("Films et nombre de films dans lesquels ont joué ni l'acteur 34, ni 98, ni 1 : \n")

movies_conditions = movies.find({"actors._id": {"$not": {"$in": ["artist:34", "artist:98", "artist:1"]}}}, {"title": 1, "actors": 1}).sort("year", -1)
count = 0
for movie in movies_conditions:
    count += 1
    print(movie)

print("\n")
print(count)

print("\n")
    
# Q : Combien de films qui n’ont pas d’attribut actors ? Combien de films dont l'attribut actors n'est pas renseigné et les afficher ?
print("Films et nombre de films qui n’ont pas d’attribut actors :")

movies_conditions = movies.find({"actors._id": {"$exists": 0}}, {'title': 1})
count = 0
for movie in movies_conditions:
    count += 1
    print(movie)

print("\n")
print(count)
```

```python
# Jointure
db = client['movies_artists']
artists = db['artists']

# Q : Donnez tous les films dont le directeur est Clint Eastwood ?

eastwood = artists.find_one({"first_name": "Clint", "last_name": "Eastwood"})

print('Films dont le directeur est Clint Eastwood :')

movies_director_4 = movies.find({"director._id": eastwood['_id']}, {"title": 1, "director._id" : 1})
for movie in movies_director_4:
    print(movie)
```

```python
# Expressions régulières ?


# Q : Tous les films dont le titre commence par 'Re' ?

print("Films dont le titre commence par 'Re' : ")
# Dans le shell mongo :
# db.movies.find({"title": /^Re/}, {"actors": null, "title": 1}).pretty()
# Resultat obtenu : 
# { "_id" : "movie:36", "title" : "Reservoir dogs", "actors" : null }

movies_re = movies.find({"title": { "$regex": '^Re' }}, {"actors": 0, "summary": 0})
for movie in movies_re:
    print(movie)

print("\n")

# Q : Tous les films dont le titre finit par 'la' ?

print("Films dont le titre fini par 'la' : ")
# Dans le shell mongo :
# db.movies.find({"title": /la$/}, {"actors": null, "title": 1}).pretty()
# Resultat obtenu : 
# { "_id" : "movie:22", "title" : "Godzilla", "actors" : null }

movies_re = movies.find({"title": { "$regex": 'la$' }}, {"actors": 0, "summary": 0})
for movie in movies_re:
    print(movie)

print("\n")

# Q : Combien de films ont été joués par l'acteur dont l'id est 147 et les afficher en se limitant au titre du film et la liste des acteurs ?

print("Films et nombre de films qui ont été joués par l'acteur dont l'id est 147 : (titre et acteurs) ")

movies_147_conditions1 = movies.find({"actors._id": "artist:147"}, {"title": 1, "actors": 1})

for movie in movies_147_conditions1:
    print(movie)

print("\n")


# Q : Retourner les films joués par l'acteur 147 sans les acteurs et sans le résumé ?

print("Films et nombre de films qui ont été joués par l'acteur dont l'id est 147 : (sans acteurs et résumé) ")

movies_147_conditions2 = movies.find({"actors._id": "artist:147"}, {"actors": 0, "summary": 0})

for movie in movies_147_conditions2:
    print(movie)
```
