```python
import pymongo
from pymongo import MongoClient
```

```python
client = MongoClient('mongodb://localhost:27017/')
db = client.gym
gymnases = db.Gymnases
sportifs = db.Sportifs

print('connexion success')
```

```python
print("Documents de la collection Sportifs : \n")

sportifs.find_one()
```

```python
# 2) On cherche le (ou les) conseiller(s) s’appelant "KERVADEC" ?

print("Le (ou les) conseillers(s) s’appelant 'KERVADEC' : \n")

conseillers = sportifs.find({'Nom' : "KERVADEC"}, {"Prenom" : 1, "Nom" : 1, "IdSportif" : 1})
for conseiller in conseillers:
    print(conseiller)
```

```python
# 3) On cherche les sportifs qui ont ou bien 32 ans ou 40 ans ?

print("Les sportifs qui ont ou bien 32 ans ou 40 ans : \n")

sportifs_32_40 = sportifs.find({"$or" : [{"Age": 32}, {"Age": 40}] })
for sportif in sportifs_32_40:
    print(sportif)
```

```python
# 4) On cherche les sportifs jouant au Basket-ball ?

print("Sportifs jouant au Basket-Ball : \n")

sportifs_basket = sportifs.find({"Sports.Jouer" : "Basket ball"})
for sportif in sportifs_basket:
    print(sportif)
    print("\n")
```

```python
# 5) On cherche les sportifs ayant au moins 32 ans, ou de sexe féminin ?

print("Sportifs ayant au moins 32 ans ou de sexe feminin : \n")

sportifs_32_femmes = sportifs.find({"$or": [{"Age":{"$gte":32}}, {"Sexe": "F"}] })

for sportif in sportifs_32_femmes:
    print(sportif)
    print('\n')
```

```python
# 6) Trouver le nombre de sportifs de sexe féminin dans la base ?

print("Nombre de sportifs de sexe féminin dans la base : \n")

sportifs_nb_femmes = sportifs.find({"Sexe": "F"}).count()

print(sportifs_nb_femmes)
```

```python
# 7) Quels sont les sportifs (identifiant, nom et prénom) qui ont entre 20 et 30 ans ?


print("Sportifs qui ont entre 20 et 30 ans : \n")

sportifs_20_30 = sportifs.find({"Age":{"$gt":20,"$lt":30}}, {'IdSportif' : 1, "Nom" : 1, "Prenom" : 1})

for sportif in sportifs_20_30:
    print(sportif)
```

```python
# 8) Quels sont les sportifs (identifiant et nom) qui pratiquent du hand ball ?


print("Sportifs pratiquant du Hand ball : \n")

sportifs_basket = sportifs.find({"Sports.Jouer" : "Hand ball"}, {'IdSportif' : 1, "Nom" : 1})
for sportif in sportifs_basket:
    print(sportif)
```

```python
# 9)    Quels sont les gymnases de “Villetaneuse” ou de “Sarcelles” qui ont une surface de plus de 400 m2 ?

print("Gymnases de “Villetaneuse” ou de “Sarcelles” qui ont une surface de plus de 400 m2 : \n")

gymnases_V_S_400 = gymnases.find({"Ville": {"$in": ["VILLETANEUSE", "SARCELLES"]}, "Surface":{"$gt":400}})

for gymnase in gymnases_V_S_400:
    print(gymnase)
    print("\n")
```

```python
# 10) Dans quels gymnases peut-on jouer au hockey le mercredi après 15H ?

print("Gymnases où on peut jouer au hockey le mercredi après 15H : \n")

gymnases_hockey_15 = gymnases.find({"Seances.Libelle" : "Hockey", "Seances.Jour" : "mercredi", "Seances.Horaire" : {"$gt":15}})
for gymnase in gymnases_hockey_15:
    print(gymnase)
    print("\n")
```

```python
# 11) Quels sportifs (identifiant et nom) ne pratiquent aucun sport ?

print("Sportifs qui ne pratiquent aucun sport : \n")

sportifs_0_sport = sportifs.find({"Sports.Jouer": {"$exists": 0}}, {'IdSportif': 1, 'Nom' : 1})

for sportif in sportifs_0_sport:
    print(sportif)
```

```python
# 12) Quels gymnases n’ont pas de séances le dimanche ?

print("Gymnases qui n’ont pas de séances le dimanche : \n")

gymnases_null_dimanche = gymnases.find({"Seances.Jour": {"$nin": ["Dimanche", "dimanche"]}})
for gymnase in gymnases_null_dimanche:
    print(gymnase)
    print("\n")
```

```python
# 13) Quels gymnases ne proposent que des séances de basket ball ou de volley ball ?

print("Gymnases qui ne proposent que des séances de basket ball ou de volley ball : \n")

gymnases_null_dimanche = gymnases.find({"$nor": [{ "Seances.Libelle": { "$ne": "Basket ball", }}, { "Seances.Libelle": { "$ne": "Volley ball" }} ]})
for gymnase in gymnases_null_dimanche:
    print(gymnase)
    print("\n")
```

```python
# 14) Quels sont les entraîneurs qui sont aussi joueurs ?

print("Entraineurs qui sont aussi des joueurs : \n")

sportifs_entraineur_joueur = sportifs.find({"Sports.Jouer" : { "$exists" : True },"Sports.Entrainer" : { "$exists" : True }})
for sportif in sportifs_entraineur_joueur : 
    print(sportif)
    print("\n")
```
